```python
import numpy as np
import gudhi as gd
import persim
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
t = np.linspace(0,1,100)

X1 = np.zeros((400, 2))

for i in range(1,5):
    X1[(i-1)*100:i*100, 0] = i*np.cos(2*np.pi*t)
    X1[(i-1)*100:i*100, 1] = i*np.sin(2*np.pi*t)
    X1[(i-1)*100:i*100, :] = X1[(i-1)*100:i*100, :] + 0.02*i*np.random.randn(100, 2)
```

```python
f,ax = plt.subplots()
ax.scatter(X1[:,0], X1[:,1])
plt.show()
```
