```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d
import gudhi as gd
```

```python
xs=np.array([0,1,2,4])
ys=np.array([3,1,4,1])
f=interp1d(xs,ys)
```

```python
xnew=np.linspace(0,4)
plt.plot(xs, ys, 'o', xnew, f(xnew), '-')
plt.show()
```

```python
def pl_prokhorov(X, Y, xs, ys, eps):
    l=0
    r=max(xs)
    f=interp1d(xs,ys)
    while (r-l>eps):
        m = (r+l)/2
        Dt = gd.bottleneck_profile(X,Y, m)
        #print(m ,Dt, f(m))
        if (Dt<=f(m)):
            r=m
        else:
            l=m
    return m
```

```python
X= np.array([[0,1],[1,2],[3,4]])
Y= np.array([[0,.5],[1,2.5],[3,5]])
xs = np.array([0,5])
ys = np.array([0,10])

pl_prokhorov(X,Y,xs,ys,0.01)
```
